Instead of looking at the whole image at once, CNNs look at small portions of the image using what we call **filters** or **kernels**. These are small matrices—say, 3×3 or 5×5 in size—that move across the image one step at a time (this step size is called the **stride**). Each filter performs a **convolution operation**, which means it multiplies its values with the pixel values it’s looking at, and sums the result to produce a single value. Repeating this across the image creates a new image called a **feature map**, which highlights certain aspects of the original image, like edges or patterns.

What’s cool is that these filters start with **random values**, but through training, the network **learns what values in the filters work best** to detect useful patterns. Early layers might learn to detect simple features like straight lines or color changes. As you stack more layers, the network begins to detect more complex features—like eyes, ears, wheels, or textures. Eventually, it forms a high-level understanding of what’s in the image.

After convolution, we usually apply an **activation function** like ReLU (Rectified Linear Unit), which introduces non-linearity. This simply replaces negative values in the feature maps with zero—helping the model learn more complex patterns.

Next, we often apply **pooling**, typically **max pooling**, which reduces the size of the feature maps. It works by taking small blocks (like 2×2 windows) and keeping only the highest value in each block. This helps the model focus on the most important features while reducing computation and overfitting.

Once the image has passed through several layers of convolution, activation, and pooling, we **flatten** the output into a long vector. This vector is then passed into a regular **fully connected neural network**, which performs the final classification. For example, if the network is trained to recognize digits (0–9), the output layer will have 10 neurons, and the one with the highest score tells us what the model thinks the ima a dog in a photo).

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision.datasets import Food101
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

* **compose**:combines multiple image transformations into one pipeline.
* **Resize** : # Resizes all images to the same size: 128×128 pixels. Food-101 images are of varying sizes, so we normalize that.
* **ToTensor** : Converts a PIL image or NumPy image to a tensor also scales pixel values from [0, 255] to [0.0, 1.0] which helps models train better.
* **normalize**: This normalizes pixel values with: mean = 0.5, std = 0.5. It converts the pixel range from [0, 1] to [-1, 1].
Why normalize?  
*Makes training faster and more stable
*Helps the optimizer converge better
*Keeps all inputs on the same scale, so no feature dominates others.

```python
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])
```

```python
# Load dataset
train_data = Food101(root='./data', split='train', download=True, transform=transform)
test_data = Food101(root='./data', split='test', download=True, transform=transform)
```

```python
train_loader = DataLoader(train_data, batch_size=32, shuffle=True, num_workers=0)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False, num_workers=0)
```

```python
for i, (images, labels) in enumerate(train_loader):
    print(f"Batch {i}: {images.shape}")
    if i >= 2:
        break  # only print first 3 batches
```

Model

```python
# CNN Model
class FoodClassifier(nn.Module):
    def __init__(self, input_shape=3, output_shape=101, hidden_units=32, hidden_units2=64):
        super().__init__()

        self.conv1 = nn.Sequential(
            nn.Conv2d(input_shape, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units2, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        self.conv2 = nn.Sequential(
            nn.Conv2d(hidden_units2, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        # Compute flattened size dynamically
        with torch.no_grad():
            sample = torch.zeros(1, input_shape, 128, 128)
            sample_out = self.conv2(self.conv1(sample))
            flat_dim = sample_out.view(1, -1).shape[1]

        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(flat_dim, 512),
            nn.ReLU(),
            nn.Linear(512, output_shape)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.fc(x)
        return x

# Initialize model
model = FoodClassifier().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
from tqdm import tqdm

epochs = 2
for epoch in range(epochs):
    model.train()
    total_loss = 0.0

    print(f"\nEpoch [{epoch+1}/{epochs}]")

    # Progress bar for batches
    for batch_idx, (images, labels) in enumerate(tqdm(train_loader, desc="Training", leave=False)):
        images, labels = images.to(device), labels.to(device)

        # Forward
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}] - Loss: {total_loss:.4f}")
```

```python
# Evaluation function
def evaluate(model, dataloader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    print(f"Test Accuracy: {100 * correct / total:.2f}%")

# Evaluate after training
evaluate(model, test_loader)

# Save the model (optional)
torch.save(model.state_dict(), "food_classifier.pth")
print("Model saved to food_classifier.pth")
```
